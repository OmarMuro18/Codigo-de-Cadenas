```python
def VCC(vec):
    ide = [ [ 1,  2, -1,  0 ], [ 0,  1,  2, -1],  [-1,  0,  1,  2],  [ 2, -1,  0,  1]]  
    caractRes = ""
    
    tamEs = len(vec)
    j = int(vec[tamEs-1])
    for x in range(tamEs-1):
        aux1 = int(vec[x])
        aux2 = int(vec[x+1])
        if (ide[aux1][aux2] == -1):
            caractRes = caractRes + "0"
        else:
            caractRes = caractRes + str(ide[aux1][aux2])
            
    aux1 = int(vec[x])
    aux2 = int(vec[x+1])
    
    if (ide[aux1][aux2]) == -1:
        caractRes = caractRes + "0"
    else:
        caractRes = caractRes + str(ide[aux1][aux2])

    return caractRes
```

```python
def F3OT(vec):
    ide = [[ 1,-1, 2,-1 ], [-1, 1,-1, 2], [ 2,-1, 1,-1],  [-1, 2,-1, 1]]  
    caractRes = ""
    
    tamEs = len(vec)
    j = int(vec[tamEs-1])
    for x in range(tamEs-1):
        aux1 = int(vec[x])
        aux2 = int(vec[x+1])
        if aux1 == aux2:
            caractRes = caractRes + "0"
        else:
            if ide[j][aux2] != -1:
                caractRes = caractRes + str(ide[j][aux2])
                
            j = aux1
    aux1 = int(vec[x])
    aux2 = int(vec[x+1])
    
    if (aux1 == aux2):
        caractRes = caractRes + "0"
    else:
        caractRes = caractRes + str(ide[j][aux2])

    return caractRes
```
